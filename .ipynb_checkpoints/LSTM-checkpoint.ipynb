```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime, timedelta as Timedelta
from alpaca_trade_api.rest import REST, TimeFrame
import torch_directml
from lumibot.brokers import Alpaca
from lumibot.backtesting import YahooDataBacktesting
from lumibot.strategies.strategy import Strategy
from finbert_utils import estimate_sentiment

# Set up the DirectML device
dml = torch_directml.device()

API_KEY = "PKWXZ2KV0T3ODZ2JBSIE" 
API_SECRET = "VJwFJNKOXAgoWAJICPgFvWxjaBEBUEvgWtK3fm1z" 
BASE_URL = "https://paper-api.alpaca.markets"

ALPACA_CREDS = {
    "API_KEY":API_KEY, 
    "API_SECRET": API_SECRET, 
    "PAPER": True
}

class LSTM(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=100, output_size=1):
        super(LSTM, self).__init__()
        self.hidden_layer_size = hidden_layer_size
        self.lstm = nn.LSTM(input_size, hidden_layer_size)
        self.linear = nn.Linear(hidden_layer_size, output_size)
        self.hidden_cell = (torch.zeros(1, 1, self.hidden_layer_size, device=dml),
                            torch.zeros(1, 1, self.hidden_layer_size, device=dml))

    def forward(self, input_seq):
        lstm_out, self.hidden_cell = self.lstm(input_seq.view(len(input_seq), 1, -1), self.hidden_cell)
        predictions = self.linear(lstm_out.view(len(input_seq), -1))
        return predictions[-1].view(-1)

class MLTrader(Strategy): 
    def initialize(self, symbol:str="SPY", cash_at_risk:float=.5): 
        self.symbol = symbol
        self.sleeptime = "24H" 
        self.last_trade = None 
        self.cash_at_risk = cash_at_risk
        self.api = REST(base_url=BASE_URL, key_id=API_KEY, secret_key=API_SECRET)
        self.model = LSTM().to(dml)  # Move the model to the DirectML device
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        self.train_model()

    def train_model(self):
        prices = self.get_historical_prices(train=True)
        scaled_prices = self.scaler.fit_transform(np.array(prices).reshape(-1, 1))

        # Ensure there is enough data for both training and validation sets
        min_data_points = 400  # Minimum data points to ensure we have enough data
        if len(scaled_prices) < min_data_points:
            raise ValueError("Not enough data points. Ensure you have at least {} data points.".format(min_data_points))

        train_size = int(len(scaled_prices) * 0.8)
        train_data = scaled_prices[:train_size]
        val_data = scaled_prices[train_size:]

        train_inout_seq = self.create_inout_sequences(train_data, 60)
        val_inout_seq = self.create_inout_sequences(val_data, 60)

        if len(val_inout_seq) < 10:  # Ensure at least 10 validation sequences
            raise ValueError("Validation set is too small. Increase the data size or adjust the train-validation split.")

        loss_function = nn.MSELoss()
        optimizer = optim.Adam(self.model.parameters(), lr=0.001)
        
        epochs = 150
        train_losses = []
        val_losses = []

        for i in range(epochs):
            train_loss = 0
            self.model.train()
            for seq, labels in train_inout_seq:
                seq, labels = seq.to(dml), labels.to(dml)  # Move the tensors to the DirectML device
                optimizer.zero_grad()
                self.model.hidden_cell = (torch.zeros(1, 1, self.model.hidden_layer_size, device=dml),
                                          torch.zeros(1, 1, self.model.hidden_layer_size, device=dml))

                y_pred = self.model(seq)

                single_loss = loss_function(y_pred, labels.view(-1))
                single_loss.backward()
                optimizer.step()
                train_loss += single_loss.item()

            train_losses.append(train_loss / len(train_inout_seq))

            val_loss = 0
            self.model.eval()
            with torch.no_grad():
                for seq, labels in val_inout_seq:
                    seq, labels = seq.to(dml), labels.to(dml)
                    self.model.hidden_cell = (torch.zeros(1, 1, self.model.hidden_layer_size, device=dml),
                                              torch.zeros(1, 1, self.model.hidden_layer_size, device=dml))
                    y_pred = self.model(seq)
                    single_loss = loss_function(y_pred, labels.view(-1))
                    val_loss += single_loss.item()

            val_losses.append(val_loss / len(val_inout_seq))

            if i % 25 == 1:
                print(f'epoch: {i:3} train loss: {train_losses[-1]:10.8f} val loss: {val_losses[-1]:10.8f}')

        # Plot the training and validation losses
        plt.figure(figsize=(10, 5))
        plt.plot(train_losses, label='Training Loss')
        plt.plot(val_losses, label='Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

    def create_inout_sequences(self, input_data, tw):
        inout_seq = []
        L = len(input_data)
        for i in range(L - tw):
            train_seq = input_data[i:i + tw]
            train_label = input_data[i + tw:i + tw + 1]
            inout_seq.append((torch.FloatTensor(train_seq), torch.FloatTensor(train_label)))
        return inout_seq

    def get_dates(self): 
        today = self.get_datetime()
        three_days_prior = today - Timedelta(days=3)
        return today.strftime('%Y-%m-%d'), three_days_prior.strftime('%Y-%m-%d')

    def get_sentiment(self): 
        today, three_days_prior = self.get_dates()
        news = self.api.get_news(symbol=self.symbol, 
                                 start=three_days_prior, 
                                 end=today) 
        news = [ev.__dict__["_raw"]["headline"] for ev in news]
        probability, sentiment = estimate_sentiment(news)
        return probability, sentiment 

    def get_historical_prices(self, train=False):
        today = self.get_datetime()
        start_date = (today - Timedelta(days=60)).strftime('%Y-%m-%d') if train else (today - Timedelta(days=30)).strftime('%Y-%m-%d')
        end_date = today.strftime('%Y-%m-%d')
        
        bars = self.api.get_bars(self.symbol, TimeFrame.Hour, start=start_date, end=end_date).df
        return bars['close'].values.tolist()

    def predict_price(self):
        prices = self.get_historical_prices()
        scaled_prices = self.scaler.transform(np.array(prices).reshape(-1, 1))
        x_test = torch.FloatTensor(scaled_prices[-60:]).view(-1).to(dml)  # Move the tensor to the DirectML device
        self.model.hidden_cell = (torch.zeros(1, 1, self.model.hidden_layer_size, device=dml),
                                  torch.zeros(1, 1, self.model.hidden_layer_size, device=dml))
        with torch.no_grad():
            predicted_price = self.model(x_test)
        predicted_price = self.scaler.inverse_transform(predicted_price.cpu().view(-1, 1))  # Move tensor to CPU before inverse transforming
        return predicted_price[0, 0]

    def position_sizing(self): 
        cash = self.get_cash() 
        last_price = self.get_last_price(self.symbol)
        quantity = round(cash * self.cash_at_risk / last_price, 0)
        return cash, last_price, quantity

    def on_trading_iteration(self):
        cash, last_price, quantity = self.position_sizing() 
        probability, sentiment = self.get_sentiment()
        predicted_price = self.predict_price()

        if cash > last_price: 
            if sentiment == "positive" and probability > .999 and predicted_price > last_price: 
                if self.last_trade == "sell": 
                    self.sell_all() 
                order = self.create_order(
                    self.symbol, 
                    quantity, 
                    "buy", 
                    type="bracket", 
                    take_profit_price=last_price*1.20, 
                    stop_loss_price=last_price*.95
                )
                self.submit_order(order) 
                self.last_trade = "buy"
            elif sentiment == "negative" and probability > .999 and predicted_price < last_price: 
                if self.last_trade == "buy": 
                    self.sell_all() 
                order = self.create_order(
                    self.symbol, 
                    quantity, 
                    "sell", 
                    type="bracket", 
                    take_profit_price=last_price*.8, 
                    stop_loss_price=last_price*1.05
                )
                self.submit_order(order) 
                self.last_trade = "sell"

start_date = datetime(2020,1,1)
end_date = datetime(2023,12,31) 
broker = Alpaca(ALPACA_CREDS) 
strategy = MLTrader(name='mlstrat', broker=broker, 
                    parameters={"symbol":"SPY", 
                                "cash_at_risk":.5})
strategy.backtest(
    YahooDataBacktesting, 
    start_date, 
    end_date, 
    parameters={"symbol":"SPY", "cash_at_risk":.5}
)
```
